```python
import matplotlib.pyplot as plt
import tensorflow.keras as keras
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import load_model

import glob
import os, os.path
from os.path import basename
import pandas as pd
import matplotlib.pyplot as plt
import PIL.Image

from PIL import Image
```

```python
classification_list =['apple', 'kaki']#, 'pumpkin', 'orange']
classification_list
```

```python
BASE_DIR = '/home/aprameyo/Weekly_exercises/Week9/imageclassifier/imageclassifier/data/'

Y_list = []    # initiate an empty list of labels
images_X=[]  # initiate an empty list of images
for class_item in classification_list: 
    for image in sorted(os.listdir(BASE_DIR + class_item)): 
        images_X.append(PIL.Image.open((os.path.join(BASE_DIR + class_item, image)))) # append image to the list
        Y_list.append(class_item)      # append labels to the list
```

```python
images_X[2]
```

```python
# Loading X

X_list = [np.array(im) for im in images_X]  # creating a list of numpy arrays of the images
X = np.array(X_list)     # create a numpy array of the list
X = X.astype('float32') / 255   ## scale the output to range 0,1 

X.shape
```

```python
# Loading Y
label_encoder = LabelEncoder()      # load LabelEncoder transform non-numerical labels to numerical labels. 
Y_num = label_encoder.fit_transform(np.array(Y_list))

# one-hot encode the output
Y = keras.utils.to_categorical(Y_num)
Y.shape
```

```python
Y
```

## Define the model

```python
model = keras.models.Sequential()

# first convolutional block
# add a convolutional layer with 32 units and a 3x3 kernel
model.add(
    keras.layers.Conv2D(32, (3,3), activation=keras.activations.relu, input_shape=(224, 224, 3))
)
# add a MaxPooling2D layer with a pool size of 2x2
model.add(
    keras.layers.MaxPooling2D((2, 2))
)
#model.add(keras.layers.BatchNormalization())

# second convolutional block
# add a convolutional layer with 64 units and a 3x3 kernel
model.add(
    keras.layers.Conv2D(64, (3,3), activation=keras.activations.relu)
)
# add a MaxPooling2D layer with a pool size of 2x2
model.add(
    keras.layers.MaxPooling2D((2, 2))
)

#model.add(keras.layers.BatchNormalization())

# third convolutional block
# add a convolutional layer with 64 units and a 3x3 kernel


# final dense block
# add a layer that flattens the output
model.add(keras.layers.Flatten())
# add a dense layer with 64 units
model.add(keras.layers.Dense(64, activation=keras.activations.relu))
# dropout layer
model.add(keras.layers.Dropout(0.7))

# add a dense output layer with 2 units and softmax activation
model.add(keras.layers.Dense(2, activation=keras.activations.softmax))
```

```python
model.summary()
```

## Compile the model

```python
# For a multi-class classification problem use categorical crossentropy
model.compile(optimizer=keras.optimizers.Adam(),
              loss=keras.losses.categorical_crossentropy,
              metrics=['accuracy'])
```

## training the model

```python
callback = keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)

# convert y to a dummy encoded matrix
history = model.fit(X, Y, 
                    epochs=30,
                    batch_size=40, 
                    verbose=2, 
                    callbacks=[callback], 
                    validation_split=0.2  # use 10% of the data for validation
)
```
